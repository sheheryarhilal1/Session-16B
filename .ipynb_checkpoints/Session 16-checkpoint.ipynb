## In Machine Learning, OVR stands for One-vs-Rest (or One-vs-All).

It is a strategy used for multi-class classification where a model is trained like this:

## Email Spam Classification

Suppose you want to classify emails into 3 classes:

1.Spam
2.Promotions
3.Important

**How OVR works:**

**You train 3 models:**

`Model 1:` Spam vs (Promotions + Important)

`Model 2:` Promotions vs (Spam + Important)

`Model 3:` Important vs (Spam + Promotions)

When a new email comes:

All 3 models give a score:

Spam model → 0.2
Promotions model → 0.6
Important model → 0.3

Highest score = Promotions (0.6)

So final prediction = Promotions

### Simple summary:

OVR means each class gets its own “yes/no detector”, and the strongest detector decides the final class.

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("ovr_real_world_categorical.csv")  # ya dataset file
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

for col in ["gender", "city", "education"]:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    
```

```python
df.head()
```

```python
df['city'].unique()
```

```python
df['education'].unique()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[["age", "income"]] = scaler.fit_transform(df[["age", "income"]])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop("target_class", axis=1)
y = df["target_class"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression

model = OneVsRestClassifier(LogisticRegression(max_iter=1000))
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```
